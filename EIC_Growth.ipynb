```python
import pandas as pd
from sklearn import linear_model
import warnings
warnings.filterwarnings('ignore')

print("Hello 👋🏻! Let me help you predict your plant growth on the end of 5 days based on your input factors: ")
print()

print("Enter 1 for Classical")
print("2 for Rock")
print("3 for No Music")
print()
msc = input("Enter the type of music(1/2/3 only)")

print("Enter 1 for Red")
print("2 for Blue")
print("3 for Red&Blue")
print()
lgt = input("Enter the colour of grow light(1/2/3 only)")
print()

df = pd.read_csv("growth_data.csv")

X = df[['Music', 'Light']]
y = df['Height(cm)']

regr = linear_model.LinearRegression()
regr.fit(X, y)

growth = regr.predict([[msc, lgt]])

rounded_growth = round(float(growth),1)
print("The expected growth of the plant is",rounded_growth,"cm.")
print()

if rounded_growth >= 5.0:
  print("The plant growth will be high!👏🏻")

elif rounded_growth >= 4.0 and rounded_growth < 5.0:
  print("The plant growth will be close to ideal!👍🏻")

else:
  print("The plant growth will be low!😟")
```
